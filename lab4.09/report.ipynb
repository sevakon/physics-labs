–°–≤–µ—Ç, –≤ –∫–æ—Ç–æ—Ä–æ–º –Ω–∞–ø—Ä–∞–≤–ª–µ–Ω–∏–µ –∫–æ–ª–µ–±–∞–Ω–∏–π —Å–≤–µ—Ç–æ–≤–æ–≥–æ –≤–µ–∫—Ç–æ—Ä–∞ –∫–∞–∫–∏–º—Ç–æ –æ–±—Ä–∞–∑–æ–º —É–ø–æ—Ä—è–¥–æ—á–µ–Ω–æ, –Ω–∞–∑—ã–≤–∞–µ—Ç—Å—è –ø–æ–ª—è—Ä–∏–∑–æ–≤–∞–Ω–Ω—ã–º. –ï—Å–ª–∏ –≤ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–µ –∫–∞–∫–∏—Ö-—Ç–æ –≤–Ω–µ—à–Ω—à–∏–π –≤–æ–∑–¥–µ–π—Å—Ç–≤–∏–π –ø–æ—è–≤–ª—è–µ—Ç—Å—è –ø—Ä–µ–∏–º—É—â–µ—Å—Ç–≤–µ–Ω–Ω–æ–µ, –Ω–æ –Ω–µ –µ–¥–∏–Ω—Å—Ç–≤–µ–Ω–Ω–æ–µ –Ω–∞–ø—Ä–∞–≤–ª–µ–Ω–∏–µ –∫–æ–ª–µ–±–∞–Ω–∏–π –≤–µ–∫—Ç–æ—Ä–∞ ùê∏, —Ç–æ —Ç–∞–∫–æ–π —Å–≤–µ—Ç –Ω–∞–∑—ã–≤–∞—é—Ç —á–∞—Å—Ç–∏—á–Ω–æ –ø–æ–ª—è—Ä–∏–∑–æ–≤–∞–Ω–Ω—ã–º.

–°–≤–µ—Ç, –≤ –∫–æ—Ç–æ—Ä–æ–º –≤–µ–∫—Ç–æ—Ä ùê∏ –∫–æ–ª–µ–±–ª–µ—Ç—Å—è —Ç–æ–ª—å–∫–æ –≤ –æ–¥–Ω–æ–º –≤—ã–¥–µ–ª–µ–Ω–Ω–æ–º –Ω–∞–ø—Ä–∞–≤–ª–µ–Ω–∏–∏, –ø–µ—Ä–ø–µ–Ω–¥–∏–∫—É–ª—è—Ä–Ω–æ–º –ª—É—á—É, –Ω–∞–∑—ã–≤–∞–µ—Ç—Å—è –ª–∏–Ω–µ–π–Ω–æ –ø–æ–ª—è—Ä–∏–∑–æ–≤–∞–Ω–Ω—ã–º

–ü–ª–æ—Å–∫–æ—Å—Ç—å, –ø—Ä–æ—Ö–æ–¥—è—â–∞—è —á–µ—Ä–µ–∑ –Ω–∞–ø—Ä–∞–≤–ª–µ–Ω–∏–µ –∫–æ–ª–µ–±–∞–Ω–∏–π —Å–≤–µ—Ç–æ–≤–æ–≥–æ
–≤–µ–∫—Ç–æ—Ä–∞ –ø–ª–æ—Å–∫–æ–ø–æ–ª—è—Ä–∏–∑–æ–≤–∞–Ω–Ω–æ–π –≤–æ–ª–Ω—ã –∏ –Ω–∞–ø—Ä–∞–≤–ª–µ–Ω–∏–µ —Ä–∞—Å–ø—Ä–æ—Å—Ç—Ä–∞–Ω–µ–Ω–∏—è —ç—Ç–æ–π –≤–æ–ª–Ω—ã, –Ω–∞–∑—ã–≤–∞–µ—Ç—Å—è –ø–ª–æ—Å–∫–æ—Å—Ç—å—é –ø–æ–ª—è—Ä–∏–∑–∞—Ü–∏–∏.

–ï—Å—Ç–µ—Å—Ç–≤–µ–Ω–Ω—ã–π —Å–≤–µ—Ç –º–æ–∂–Ω–æ –ø—Ä–µ–≤—Ä–∞—Ç–∏—Ç—å –≤ –ø–ª–æ—Å–∫–æ–ø–æ–ª—è—Ä–∏–∑–æ–≤–∞–Ω—ã–π —Å –ø–æ–º–æ—â—å—é —É—Å—Ç—Ä–æ–π—Å—Ç–≤–∞, –∫–æ—Ç–æ—Ä–æ–µ –Ω–∞–∑—ã–≤–∞—é—Ç –ø–æ–ª—è—Ä–∏–∑–∞—Ç–æ—Ä–æ–º.

#### –ó–∞–∫–æ–Ω –ú–∞–ª—é—Å–∞
$I = I_0 \cos^2 \phi$, –≥–¥–µ $\phi$ - —É–≥–æ–ª –º–µ–∂–¥—É –Ω–∞–ø—Ä–∞–≤–ª–µ–Ω–∏—è–º–∏ –ø—Ä–æ–ø—É—Å–∫–∞–Ω–∏—è –ø–æ–ª—è—Ä–∏–∑–∞—Ç–æ—Ä–∞ –∏
–Ω–∞–ø—Ä–∞–≤–ª–µ–Ω–∏–µ–º –∫–æ–ª–µ–±–∞–Ω–∏–π —Å–≤–µ—Ç–æ–≤–æ–≥–æ –≤–µ–∫—Ç–æ—Ä–∞, $I_0$ - –∏–Ω—Ç–µ–Ω—Å–∏–≤–Ω–æ—Å—Ç—å –ø–∞–¥–∞—é—â–µ–≥–æ –Ω–∞ –ø–æ–ª—è—Ä–∏–∑–∞—Ç–æ—Ä —Å–≤–µ—Ç–∞, $I$ - 
–∏–Ω—Ç–µ–Ω—Å–∏–≤–Ω–æ—Å—Ç—å –ø—Ä–æ—à–µ–¥—à–µ–≥–æ —á–µ—Ä–µ–∑ –ø–æ–ª—è—Ä–∏–∑–∞—Ç–æ—Ä —Å–≤–µ—Ç–∞. 

#### –§–æ—Ä–º—É–ª—ã –§—Ä–µ–Ω–µ–ª—è
–ö–æ–ª–∏—á–µ—Å—Ç–≤–µ–Ω–Ω—ã–µ —Å–æ–æ—Ç–Ω–æ—à–µ–Ω–∏—è –º–µ–∂–¥—É –∏–Ω—Ç–µ–Ω—Å–∏–≤–Ω–æ—Å—Ç—è–º–∏ –ø–∞–¥–∞—é—â–µ–≥–æ, –æ—Ç—Ä–∞–∂–µ–Ω–Ω–æ–≥–æ –∏ –ø—Ä–æ—à–µ–¥—à–µ–≥–æ —Å–≤–µ—Ç–∞ –º–æ–∂–Ω–æ –ø–æ–ª—É—á–∏—Ç—å –∏–∑ –≥—Ä–∞–Ω–∏—á–Ω—ã—Ö —É—Å–ª–æ–≤–∏–π –¥–ª—è –Ω–∞–ø—Ä—è–∂–µ–Ω–Ω–æ—Å—Ç–µ–π —ç–ª–µ–∫—Ç—Ä–∏—á–µ—Å–∫–æ–≥–æ –∏ –º–∞–≥–Ω–∏—Ç–Ω–æ–≥–æ –ø–æ–ª–µ–π –Ω–∞ –≥—Ä–∞–Ω–∏—Ü–µ —Ä–∞–∑–¥–µ–ª–∞ –¥–≤—É—Ö –¥–∏—ç–ª–µ–∫—Ç—Ä–∏–∫–æ–≤. –≠—Ç–∏ —Å–æ–æ—Ç–Ω–æ—à–µ–Ω–∏—è –±—ã–ª–∏ –ø–æ–ª—É—á–µ–Ω—ã –§—Ä–µ–Ω–µ–ª–µ–º –∏ –Ω–æ—Å—è—Ç –Ω–∞–∑–≤–∞–Ω–∏–µ —Ñ–æ—Ä–º—É–ª –§—Ä–µ–Ω–µ–ª—è.

–ö–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –æ—Ç—Ä–∞–∂–µ–Ω–∏—è $R = \frac{I_{refl}}{I_{fal}}$ 

–ï—Å–ª–∏ —É—á–∏—Ç—ã–≤–∞—Ç—å, —á—Ç–æ –∏–Ω—Ç–µ–Ω—Å–∏–≤–Ω–æ—Å—Ç—å –ø—Ä–æ–ø–æ—Ä—Ü–∏–æ–Ω–∞–ª—å–Ω–∞ –∫–≤–∞–¥—Ä–∞—Ç—É –∞–º–ø–ª–∏—Ç—É–¥—ã, –≤—ã—Ä–∞–∑–∏–º –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç—ã –æ—Ç—Ä–∞–∂–µ–Ω–∏—è –¥–ª—è –ø—Ä–æ–¥–æ–ª—å–Ω–æ–π –∏ –ø–æ–ø–µ—Ä–µ—á–Ω–æ–π —Å–æ—Å—Ç–∞–≤–ª—è—é—â–∏—Ö:

$R^{\parallel} = \left( \frac{E_{refl}^{\parallel}}{E_{fal}^{\parallel}} \right)^2 = \frac{\tan^2 (\phi - \psi)}{\tan^2 (\phi + \psi)}$

$R^{\bot} = \left( \frac{E_{refl}^{\bot}}{E_{fal}^{\bot}} \right)^2 = \frac{\sin^2 (\phi - \psi)}{\sin^2 (\phi + \psi)}$

–°—Ç–µ–ø–µ–Ω—å –ø–æ–ª—è—Ä–∏–∑–∞—Ü–∏–∏ —Å–≤–µ—Ç–∞ –∑–∞–≤–∏—Å–∏—Ç –æ—Ç —É–≥–ª–∞ –ø–∞–¥–µ–Ω–∏—è –ª—É—á–µ–π –∏ –ø–æ–∫–∞–∑–∞—Ç–µ–ª—è –ø—Ä–µ–ª–æ–º–ª–µ–Ω–∏—è. –®–æ—Ç–ª–∞–Ω–¥—Å–∫–∏–π —Ñ–∏–∑–∏–∫ –î. –ë—Ä—é—Å—Ç–µ—Ä (1781-1868) —É—Å—Ç–∞–Ω–æ–≤–∏–ª –∑–∞–∫–æ–Ω, —Å–æ–≥–ª–∞—Å–Ω–æ –∫–æ—Ç–æ—Ä–æ–º—É –ø—Ä–∏ —É–≥–ª–µ –ø–∞–¥–µ–Ω–∏—è $\phi_{br}$ (—É–≥–æ–ª –ë—Ä—é—Å—Ç–µ—Ä–∞), –æ–ø—Ä–µ–¥–µ–ª—è–µ–º–æ–º —Å–æ–æ—Ç–Ω–æ—à–µ–Ω–∏–µ–º: 

$\tan \phi_{br} = n_{21}$, 

–≥–¥–µ $n_{21}$ - —ç—Ç–æ –ø–æ–∫–∞–∑–∞—Ç–µ–ª—å –ø—Ä–µ–ª–æ–º–ª–µ–Ω–∏—è –≤—Ç–æ—Ä–æ–π —Å—Ä–µ–¥—ã –æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω–æ –ø–µ—Ä–≤–æ–π, –æ—Ç—Ä–∞–∂—ë–Ω–Ω—ã–π –ª—É—á —è–≤–ª—è–µ—Ç—Å—è –ø–ª–æ—Å–∫–æ–ø–æ–ª—è—Ä–∏–∑–æ–≤–∞–Ω–Ω—ã–º (—Å–æ–¥–µ—Ä–∂–∏—Ç —Ç–æ–ª—å–∫–æ –∫–æ–ª–µ–±–∞–Ω–∏—è, –ø–µ—Ä–ø–µ–Ω–¥–∏–∫—É–ª—è—Ä–Ω—ã–µ –ø–ª–æ—Å–∫–æ—Å—Ç–∏ –ø–∞–¥–µ–Ω–∏—è).

–ü—Ä–µ–ª–æ–º–ª—ë–Ω–Ω—ã–π –∂–µ –ª—É—á, –ø—Ä–∏ —É–≥–ª–µ –ø–∞–¥–µ–Ω–∏—è $\phi_{br}$ –ø–æ–ª—è—Ä–∏–∑—É–µ—Ç—Å—è –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ, –Ω–æ –Ω–µ –ø–æ–ª–Ω–æ—Å—Ç—å—é. –ï—Å–ª–∏ —Å–≤–µ—Ç –ø–∞–¥–∞–µ—Ç –Ω–∞ –≥—Ä–∞–Ω–∏—Ü—É —Ä–∞–∑–¥–µ–ª–∞ –ø–æ–¥ —É–≥–ª–æ–º –ë—Ä—é—Å—Ç–µ—Ä–∞, —Ç–æ –æ—Ç—Ä–∞–∂—ë–Ω–Ω—ã–π –∏ –ø—Ä–µ–ª–æ–º–ª—ë–Ω–Ω—ã–π –ª—É—á–∏ –≤–∑–∞–∏–º–Ω–æ –ø–µ—Ä–ø–µ–Ω–¥–∏–∫—É–ª—è—Ä–Ω—ã.

$\tan \phi_{br} = \frac{\sin \phi_{br}}{\cos \phi_{br}}$; $\space \space n_{21} = \frac{\sin \phi_{br}}{\sin \psi}$ 

$\cos \phi_{br} = \sin \psi$

$\phi_{br} + \psi = \frac{\pi}{2}$

–°—Ç–µ–ø–µ–Ω—å –ø–æ–ª—è—Ä–∏–∑–∞—Ü–∏–∏ –ø—Ä–µ–ª–æ–º–ª–µ–Ω–Ω–æ–≥–æ —Å–≤–µ—Ç–∞ –º–æ–∂–µ—Ç –±—ã—Ç—å –∑–Ω–∞—á–∏—Ç–µ–ª—å–Ω–æ –ø–æ–≤—ã—à–µ–Ω–∞ –º–Ω–æ–≥–æ–∫—Ä–∞—Ç–Ω—ã–º –ø—Ä–µ–ª–æ–º–ª–µ–Ω–∏–µ–º –ø—Ä–∏ —É—Å–ª–æ–≤–∏–∏ –ø–∞–¥–µ–Ω–∏—è —Å–≤–µ—Ç–∞ –∫–∞–∂–¥—ã–π —Ä–∞–∑ –Ω–∞ –≥—Ä–∞–Ω–∏—Ü—É —Ä–∞–∑–¥–µ–ª–∞ –ø–æ–¥ —É–≥–ª–æ–º –ë—Ä—é—Å—Ç–µ—Ä–∞.
–ï—Å–ª–∏, –Ω–∞–ø—Ä–∏–º–µ—Ä, –¥–ª—è —Å—Ç–µ–∫–ª–∞ ùëõ = 1, 53, —Å—Ç–µ–ø–µ–Ω—å –ø–æ–ª—è—Ä–∏–∑–∞—Ü–∏–∏ –ø—Ä–µ–ª–æ–º–ª–µ–Ω–Ω–æ–≥–æ –ª—É—á–∞ —Å–æ—Å—Ç–∞–≤–ª—è–µ—Ç ‚âà 15%, —Ç–æ –ø–æ—Å–ª–µ –ø—Ä–µ–ª–æ–º–ª–µ–Ω–∏—è –Ω–∞ 8-10 –Ω–∞–ª–æ–∂–µ–Ω–Ω—ã—Ö –¥—Ä—É–≥ –Ω–∞ –¥—Ä—É–≥–∞ —Å—Ç–µ–∫–ª—è–Ω–Ω—ã—Ö –ø–ª–∞—Å—Ç–∏–Ω–∫–∞—Ö, –≤—ã—à–µ–¥—à–∏–π –∏–∑
—Ç–∞–∫–æ–π —Å–∏—Å—Ç–µ–º—ã —Å–≤–µ—Ç –±—É–¥–µ—Ç –ø—Ä–∞–∫—Ç–∏—á–µ—Å–∫–∏ –ø–æ–ª–Ω–æ—Å—Ç—å—é –ø–æ–ª—è—Ä–∏–∑–æ–≤–∞–Ω–Ω—ã–º.
–¢–∞–∫–∞—è —Å–æ–≤–æ–∫—É–ø–Ω–æ—Å—Ç—å –ø–ª–∞—Å—Ç–∏–Ω–æ–∫ –Ω–∞–∑—ã–≤–∞–µ—Ç—Å—è —Å—Ç–æ–ø–æ–π –°—Ç–æ–ª–µ—Ç–æ–≤–∞. –°—Ç–æ–ø–∞ –º–æ–∂–µ—Ç —Å–ª—É–∂–∏—Ç—å –¥–ª—è –∞–Ω–∞–ª–∏–∑–∞ –ø–æ–ª—è—Ä–∏–∑–æ–≤–∞–Ω–Ω–æ–≥–æ —Å–≤–µ—Ç–∞ –∫–∞–∫ –ø—Ä–∏ –µ–≥–æ
–æ—Ç—Ä–∞–∂–µ–Ω–∏–∏, —Ç–∞–∫ –∏ –ø—Ä–∏ –µ–≥–æ –ø—Ä–µ–ª–æ–º–ª–µ–Ω–∏–∏.

## –¶–µ–ª—å —Ä–∞–±–æ—Ç—ã 
–ò–∑—É—á–µ–Ω–∏–µ –ø–æ–ª—è—Ä–∏–∑–æ–≤–∞–Ω–Ω–æ–≥–æ —Å–≤–µ—Ç–∞ –∏ –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ø–æ–∫–∞–∑–∞—Ç–µ–ª–µ–π –ø—Ä–µ–ª–æ–º–ª–µ–Ω–∏—è.

## –†–∞–±–æ—á–∏–µ —Ñ–æ—Ä–º—É–ª—ã –∏ –∏—Å—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ

#### –ò–∑—É—á–µ–Ω–∏–µ –ø–æ–ª—è—Ä–∏–∑–æ–≤–∞–Ω–Ω–æ–≥–æ —Å–≤–µ—Ç–∞ —Å –ø–æ–º–æ—â—å—é —Å—Ç–æ–ø—ã –°—Ç–æ–ª–µ—Ç–æ–≤–∞

```python
import sympy
import scipy
import numpy as np
import pandas as pd
from scipy.signal import argrelextrema
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (10,5)
%matplotlib inline
```

## –†–µ–∑—É–ª—å—Ç–∞—Ç—ã –∏–∑–º–µ—Ä–µ–Ω–∏–π –∏ —Ä–∞—Å—á–µ—Ç—ã

### –≠–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç–∞–ª—å–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –ø—Ä–∏ –∏–∑—É—á–µ–Ω–∏–µ –∑–∞–∫–æ–Ω–∞ –ú–∞–ª—é—Å–∞
$\alpha$ - –≥—Ä–∞–¥—É—Å —É–≥–ª–∞ –ø–æ–≤–æ—Ä–æ—Ç–∞ –∞–Ω–∞–ª–∏–∑–∞—Ç–æ—Ä–∞, $I$ - –∏–Ω—Ç–µ–Ω—Å–∏–≤–Ω–æ—Å—Ç—å —Å–≤–µ—Ç–∞ –Ω–∞ –∞–Ω–∞–ª–∏–∑–∞—Ç–æ—Ä–µ (—Ñ–æ—Ç–æ—Ç–æ–∫ —á–µ—Ä–µ–∑ —Ñ–æ—Ç–æ—Ä–µ–≥–µ—Å—Ç—Ä–∏—Ä—É—é—â–µ–µ —É—Å—Ç—Ä–æ–π—Å—Ç–≤–æ)

```python
df1 = pd.DataFrame({
    'Œ±, –≥—Ä–∞–¥—É—Å—ã': [i * 10 for i in range(19)],
    '$I$, –º–∫–ê': 
        [72.3, 71.3, 65.8, 60.3, 50.1, 39.4, 26.5, 
         14.7, 6.5, 2.5, 4.7, 12.3, 24.5, 36.8, 47.4, 
         56.8, 64.4, 69.1, 70.4]
})

df1['cos Œ±'] = np.cos(df1['Œ±, –≥—Ä–∞–¥—É—Å—ã'] * np.pi / 180)
df1['$\cos^2 Œ±$'] = np.cos(df1['Œ±, –≥—Ä–∞–¥—É—Å—ã'] * np.pi / 180) ** 2

df1
```
