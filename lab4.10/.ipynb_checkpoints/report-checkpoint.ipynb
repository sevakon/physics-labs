$I_{o} = 1.505$ - –æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω–∞—è –∏–Ω—Ç–µ–Ω—Å–∏–≤–Ω–æ—Å—Ç—å –ª–∞–∑–µ—Ä–∞ $I_{–ø}$, –Ω–µ –æ—Å–ª–∞–±–ª–µ–Ω–Ω–∞—è –ø–æ–ª—è—Ä–∏–∑–∞—Ç–æ—Ä–æ–º

$I^\prime_{o} = 1.505$ - –æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω–∞—è –∏–Ω—Ç–µ–Ω—Å–∏–≤–Ω–æ—Å—Ç—å –∏—Å—Ç–æ—á–Ω–∏–∫–∞ –±–µ–ª–æ–≥–æ —Å–≤–µ—Ç–∞, –Ω–µ –æ—Å–ª–∞–±–ª–µ–Ω–Ω–∞—è –ø–æ–ª—è—Ä–∏–∑–∞—Ç–æ—Ä–æ–º

$I^\prime = 0.450$ - –æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω–∞—è –∏–Ω—Ç–µ–Ω—Å–∏–≤–Ω–æ—Å—Ç—å —Å–≤–µ—Ç–∞, –ø—Ä–æ—à–µ–¥—à–µ–≥–æ —á–µ—Ä–µ–∑ –ø–æ–ª—è—Ä–∏–∑–∞—Ç–æ—Ä

**–ò–∑–ª—É—á–µ–Ω–∏–µ –ª–∞–∑–µ—Ä–∞**

```python
I_o = 1.505
I_o_prime = 1.505
I_prime = 0.450

lazer_df = pd.DataFrame({
    '$\phi$, –≥—Ä–∞–¥—É—Å—ã': [i * 10 for i in range(16)],
    '$I_1$': [0.826, 0.776, 0.671, 0.604, 0.469, 0.321, 0.206, 0.107, 
              0.036, 0.003, 0.018, 0.069, 0.176, 0.274, 0.405, 0.495],
    '$I_2$': [0.835, 0.768, 0.703, 0.6, 0.479, 0.316, 0.192, 0.11, 
              0.029, 0.004, 0.011, 0.076, 0.145, 0.263, 0.353, 0.513],
})

lazer_df['$I_{mean}$'] = (lazer_df['$I_1$'] + lazer_df['$I_2$']) / 2

lazer_df
```

–≥–¥—É $\alpha$ - —É–≥–æ–ª –ø–æ–≤–æ—Ä–æ—Ç–∞ –∞–Ω–∞–ª–∏–∑–∞—Ç–æ—Ä–∞, $I_1$ –∏ $I_2$ - –∏–Ω—Ç–µ–Ω—Å–∏–≤–Ω–æ—Å—Ç–∏ –ª—É—á–∞, –ø—Ä–æ—à–µ–¥—à–µ–≥–æ
—á–µ—Ä–µ–∑ –∞–Ω–∞–ª–∏–∑–∞—Ç–æ—Ä –¥–ª—è –¥–≤—É—Ö —ç–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç–æ–≤.    
$I_{mean}$ - —Å—Ä–µ–¥–Ω–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –¥–≤—É—Ö —ç–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç–æ–≤. 

```python
def polarization_degree(i_max, i_min):
    degree = (i_max - i_min) / (i_max + i_min)
    return degree
```

```python
print('–°—Ç–µ–ø–µ–Ω—å –ø–æ–ª—è—Ä–∏–∑–∞—Ü–∏–∏ –ª–∞–∑–µ—Ä–Ω–æ–≥–æ –∏–∑–ª—É—á–µ–Ω–∏—è: {:.3f}'.format(polarization_degree(
    max(lazer_df['$I_{mean}$']), min(lazer_df['$I_{mean}$']))))
```

**–ù–∞–π–¥–µ–º –º–∞–∫—Å–∏–º–∞–ª—å–Ω—É—é –∏–Ω—Ç–µ–Ω—Å–∏–≤–Ω–Ω–æ—Å—Ç—å, —á—Ç–æ–±—ã –≤ –¥–∞–ª—å–Ω–µ–π—à–µ–º –ø–æ–ª—É—á–∏—Ç—å –æ—Ç–Ω–æ—Å–∏—Ç–µ–ª–Ω—ã–µ –∏–Ω—Ç–µ–Ω—Å–∏–≤–Ω–æ—Å—Ç–∏.**

```python
I_max = max(max(lazer_df['$I_1$']), max(lazer_df['$I_2$']))
print('I max: {}'.format(I_max))

I_min = min(min(lazer_df['$I_1$']), min(lazer_df['$I_2$']))
print('I min: {}'.format(I_min))

phi_max = 0
print('Phi max: {}'.format(phi_max))
```

**–ü–æ—Å—Ç—Ä–æ–∏–º —Ç–∞–±–ª–∏—Ü—É —Å –æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω—ã–º–∏ –∑–Ω–∞—á–µ–Ω–∏—è–º–∏:**

```python
lazer_rel_df = pd.DataFrame({
    '$\phi$, –≥—Ä–∞–¥—É—Å—ã': [i * 10 for i in range(16)]
})

lazer_rel_df['$I_{rel1}$'] = lazer_df['$I_1$'] / I_max
lazer_rel_df['$I_{rel2}$'] = lazer_df['$I_2$'] / I_max
lazer_rel_df['$I_{rel}$'] = (lazer_rel_df['$I_{rel1}$'] + lazer_rel_df['$I_{rel2}$']) / 2

lazer_rel_df
```

**–ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –≥—Ä–∞—Ñ–∏–∫–∞ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –Ω–æ—Ä–º–∏—Ä–æ–≤–∞–Ω–Ω–æ–π –∏–Ω—Ç–µ–Ω—Å–∏–≤–Ω–æ—Å—Ç–∏ $I_{–æ—Ç–Ω}$ –æ—Ç —É–≥–ª–∞ $\phi$ –ø–æ–≤–æ—Ä–æ—Ç–∞
–ø–æ–ª—è—Ä–æ–∏–¥–∞ –≤ –ø–æ–ª—è—Ä–Ω—ã—Ö –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–∞—Ö; –≥—Ä–∞—Ñ–∏–∫–∞ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ $\cos^2(\phi - \phi_m)$ –æ—Ç —É–≥–ª–∞ $\phi$ –ø–æ–≤–æ—Ä–æ—Ç–∞ –ø–æ–ª—è—Ä–æ–∏–¥–∞**

```python
fig, ax = plt.subplots(figsize=(20, 5))
ax.set_title('–ó–∞–≤–∏—Å–∏–º–æ—Å—Ç—å –Ω–æ—Ä–º–∏—Ä–æ–≤–∞–Ω–Ω–æ–π –∏–Ω—Ç–µ–Ω—Å–∏–≤–Ω–æ—Å—Ç–∏ –æ—Ç —É–≥–ª–∞ –ø–æ–≤–æ—Ä–æ—Ç–∞')

ax.scatter(lazer_rel_df['$\phi$, –≥—Ä–∞–¥—É—Å—ã'], lazer_rel_df['$I_{rel}$'], c='r')
ax.plot(lazer_rel_df['$\phi$, –≥—Ä–∞–¥—É—Å—ã'], lazer_rel_df['$I_{rel}$'], 'r--', label='$I_{–æ—Ç–Ω}$')

ax.scatter(lazer_rel_df['$\phi$, –≥—Ä–∞–¥—É—Å—ã'], (np.cos((lazer_rel_df['$\phi$, –≥—Ä–∞–¥—É—Å—ã'] - phi_max)* np.pi / 180)) ** 2, c='b')
ax.plot(lazer_rel_df['$\phi$, –≥—Ä–∞–¥—É—Å—ã'], (np.cos((lazer_rel_df['$\phi$, –≥—Ä–∞–¥—É—Å—ã'] - phi_max)* np.pi / 180)) ** 2, 'b--', label='$\cos^2(\phi - \phi_m)$')

ax.legend()

plt.show()
```

**–ò—Å—Ö–æ–¥—è –∏–∑ –≥—Ä–∞—Ñ–∏–∫–∞ –≤—ã—à–µ, –ø–æ—Å–∫–æ–ª—å–∫—É –æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω–∞—è –∏–Ω—Ç–µ–Ω—Å–∏–≤–Ω–æ—Å—Ç—å –ª–∞–∑–µ—Ä–Ω–æ–≥–æ –∏–∑–ª—É—á–µ–Ω–∏—è –æ—á–µ–Ω—å –ø–æ—Ö–æ–∂–∞ –Ω–∞ –≥—Ä–∞—Ñ–∏–∫  $\cos^2(\phi - \phi_m)$, –º–æ–∂–Ω–æ —É—Ç–≤–µ—Ä–∂–¥–∞—Ç—å, —á—Ç–æ –ª–∞–∑–µ—Ä–Ω–æ–µ –∏–∑–ª—É—á–µ–Ω–∏–µ –æ–±–ª–∞–¥–∞–µ—Ç –ª–∏–Ω–µ–π–Ω—ã–º –≤–∏–¥–æ–º –ø–æ–ª—è—Ä–∏–∑–∞—Ü–∏–∏**

**–ù–∞–π–¥–µ–º –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç—ã –ø—Ä–æ–ø—É—Å–∫–∞–Ω–∏—è –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–Ω–æ–≥–æ –ø–æ–ª—è—Ä–æ–∏–¥–∞ –¥–ª—è –ø–∞—Ä–∞–ª–ª–µ–ª—å–Ω–æ–π –∏
–ø–µ—Ä–ø–µ–Ω–¥–∏–∫—É–ª—è—Ä–Ω–æ–π –æ—Ä–∏–µ–Ω—Ç–∞—Ü–∏–∏ –µ–≥–æ –ø–ª–æ—Å–∫–æ—Å—Ç–∏ –ø—Ä–æ–ø—É—Å–∫–∞–Ω–∏—è**

```python
K_parallel = I_max / I_o
K_bot = I_min / I_o

print('–ö–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç—ã: K parallel: {:.5f}, K_bot: {:.5f}'.format(K_parallel, K_bot))
```

```python
df2 = pd.DataFrame({
    '$\phi$, –≥—Ä–∞–¥—É—Å—ã': [i for i in range(30, 65, 2)],
    '$I_1$': [1.076, 1.073, 1.071, 1.066, 1.063, 1.061, 1.059, 1.055, 1.051, 
              1.046, 1.038, 1.028, 1.018, 1.006, 0.988, 0.972, 0.955, 0.926],
    '$I_2$': [1.069, 1.063, 1.061, 1.058, 1.057, 1.054, 1.054, 1.051, 1.041, 
          1.043, 1.04, 1.028, 1.014, 0.999, 0.986, 0.976, 0.958, 0.926]
})

phi_br = 59
i_max = 0.568
i_min = 0.448

df2
```

–≥–¥—É $\phi$ - —É–≥–æ–ª –Ω–∞–∫–ª–æ–Ω–∞ —Å—Ç–µ–∫–ª—è–Ω–Ω–æ–π –ø–ª–∞—Å—Ç–∏–Ω–∫–∏, $I_1$ –∏ $I_2$ - –∏–Ω—Ç–µ–Ω—Å–∏–≤–Ω–æ—Å—Ç—å –ø—Ä–æ—à–µ–¥—à–µ–≥–æ —Å–≤–µ—Ç–∞ –≤ –ø—Ä—è–º–æ–º –∏ –æ–±—Ä–∞—Ç–Ω–æ–º –Ω–∞–ø—Ä–∞–≤–ª–µ–Ω–∏—è—Ö, —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–µ–Ω–Ω–æ. 

```python
n_21 = np.tan(phi_br * np.pi / 180)
print('–ü–æ–∫–∞–∑–∞—Ç–µ–ª—å –ø—Ä–µ–ª–æ–º–ª–µ–Ω–∏—è –≤—Ç–æ—Ä–æ–π —Å—Ä–µ–¥—ã –æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω–æ –ø–µ—Ä–≤–æ–π: {:.3f}'.format(n_21))
```

```python
p1 = polarization_degree(max(max(df2['$I_1$']), max(df2['$I_2$'])), min(min(df2['$I_1$']), min(df2['$I_2$'])))
print('C—Ç–µ–ø–µ–Ω—å –ø–æ–ª—è—Ä–∏–∑–∞—Ü–∏–∏ —Å–≤–µ—Ç–∞ –ø–æ—Å–ª–µ –ø—Ä–æ—Ö–æ–∂–¥–µ–Ω–∏—è –µ–≥–æ —á–µ—Ä–µ–∑ –ø–ª–∞—Å—Ç–∏–Ω–∫—É: {:.3f}'.format(p1))
```

```python
def partial_polarization(n):
    degree = (n ** 2 - 1) ** 2 / (2 * (n ** 2 + 1) ** 2 - (n ** 2 - 1) ** 2)
    return degree
```

```python
print('C—Ç–µ–ø–µ–Ω—å –ø–æ–ª—è—Ä–∏–∑–∞—Ü–∏–∏ —Å–≤–µ—Ç–∞ –ø–æ—Å–ª–µ –ø—Ä–æ—Ö–æ–∂–¥–µ–Ω–∏—è –µ–≥–æ —á–µ—Ä–µ–∑ –ø–ª–∞—Å—Ç–∏–Ω–∫—É —Ç–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–∞—è: {:.3f}'.format(partial_polarization(n_21)))
```

```python
p3 = polarization_degree(i_max, i_min)
print('C—Ç–µ–ø–µ–Ω—å –ø–æ–ª—è—Ä–∏–∑–∞—Ü–∏–∏ –±–µ–ª–æ–≥–æ —Å–≤–µ—Ç–∞ –ø–æ—Å–ª–µ –ø—Ä–æ—Ö–æ–∂–¥–µ–Ω–∏—è –µ–≥–æ —á–µ—Ä–µ–∑ –ø–ª–∞—Å—Ç–∏–Ω–∫—É: {:.3f}'.format(p3))
```
